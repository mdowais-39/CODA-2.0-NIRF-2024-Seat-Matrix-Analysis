COMPARISION OF GEN-EWS AND ST CATEGORY BY MERGING THEIR DATA.

```python
general_seats =df1.groupby('Institute Name')['GEN-EWS'].sum().reset_index()
st_seats =df1.groupby('Institute Name')['ST'].sum().reset_index()
```

```python
merged_df=pd.merge(general_seats,st_seats,on='Institute Name',suffixes=('_General','_ST'))
#PLOT A GROUPEDBAR CHART
plt.figure(figsize=(25, 20))
bar_width = 0.35  # Set bar width
institutes = merged_df['Institute Name']
index = range(len(institutes))

# Create the bars for General and ST categories
plt.bar(index, merged_df['GEN-EWS'], width=bar_width, label='General')
plt.bar([i + bar_width for i in index], merged_df['ST'], width=bar_width, label='ST')

# Add labels and title
plt.xlabel('Institute', fontsize=12)
plt.ylabel('Number of Seats', fontsize=12)
plt.title('Comparison of Seats for General and ST Categories', fontsize=15)
plt.xticks([i + bar_width/2 for i in index], institutes, rotation=90)

# Add legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

THIS SHOWS THE  NUMBER OF SEATS IN VARIOUS INSTITUTES VARIES GREATLY

FROM THIS, WE ALSO CONFIRM THAT THE SEATS IN THESE COLLEGES FOR CATEGORIES LIKE GEN AND ST ARE ALMOST OF NO SIGNIFICANCE. IN FACT,SIKKIM HAS A NEGLIGIBLE NUMBER OF SEATS FOR GEN CATEGORY
